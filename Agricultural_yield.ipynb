```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("samuelotiattakorah/agriculture-crop-yield")

print("Path to dataset files:", path)
import pandas as pd
```

```python
df= pd.read_csv(path+"/crop_yield.csv")
df.head()
```

```python
print(df.dtypes)
numeric_only = df.select_dtypes(exclude=['object']).columns
print(numeric_only)
```

```python
print(df.isnull().sum())
```

```python
df=df.dropna()
df
```

```python
print(df.duplicated().sum())
```

```python
df['Fertilizer_Used']= df['Fertilizer_Used'].astype(int)
df['irrigation_Used']= df['Irrigation_Used'].astype(int)
```

```python
df['Fertilizer_Used']
```

```python
df.columns
```

```python
df=pd.get_dummies(df, columns=['Region', 'Soil_Type', 'Crop','Weather_Condition'], drop_first=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.info)
print(df.head)
```

```python
x=df.drop(columns=['Yield_tons_per_hectare'])
y=df['Yield_tons_per_hectare']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
print(f"Training size:{x_train.shape}, Test size:{x_test.shape}")
```

```python
from xgboost import XGBRegressor as RandomForest
from sklearn.model_selection import GridSearchCV
params={
    'n_estimators': [100, 200, 300],
    'max_depth': [6, 10, 20],
    'learning_rate': [0.01, 0.05, 0.08],
    'subsample': [0.7, 0.8, 0.9],
    'random_state': [42]
}

model = RandomForest(n_estimators=100, max_depth=10, learning_rate=0.05, subsample=0.8, random_state=42)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test, y_pred)
r2= r2_score(y_test, y_pred)

print(f"Mean_Absolute_Error:{mae:.4f}")
print(f"Mean_Squared_Error:{mse:.4f}")
print(f"R^2:{r2:.4f}")
```
